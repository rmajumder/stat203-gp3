### 3. Model Creation

#### 3.1 Model 1 - Key Interest Variables
First we will make a model based solely on the key variables.

```r
options(repr.plot.width=10, repr.plot.height=5)
model1 <- lm(crmrte ~  prbarr + prbconv + density + taxpc  + pctymle, data = cdata)
coeftest(model1, vcov = vcovHC)
plot(model1, which = 1)
plot(model1, which = 5)
```

```r
paste("Model Adjusted R-Sq - ", summary(model1)$adj.r.squared)
paste("AIC Score: ", AIC(model1))
paste("Covariation of coefficients - ")
diag(vcov(model1))
```

#### 3.1.1 Model 1 - Interpretation 

1. Low Residuals Median, coefficients, variation of coefficients and RSE
2. Significantly high Adjusted R-squared values - 0.6841 <br>
High Adjusted R-squared value implies goodness of fit, although it may be inflated due to the number of inputs.
3. Highest AIC score: -561
4. One value more than 1 cook's distance. A few close to $\frac {1}{2}$ cook's distance. <br>
Two observations can be considered as outliers and those may influence the estimation (25, 51).
5. Density, prbconv and pctymle variables are statistically significant.
6. The model efficiency is high since the coefficient variations or robust standard error values are low. It also means the estimators are consistent around the regression line. 
7. In the residual vs fitted plot, except the small upward pull in the left side, the spline curve is almost perfectly aligned with the fitted line. The reason behind the small devation is minimal number of data points in that value range. So, it can be concluded that there is no major omitted variable bias for the model1.


#### 3.2 Model 2 - Key Variables + Covariates

#### 3.2.1 Adjustments after Model1 results

1. Removing two outliers 
2. Removing statistically insignificant variable, wage. 
3. Introducing new variables which have practical significance in determining crime rates:<br> 
    -- Wage, Police Per Capita, and Mix <br>
    
Note - These variables have a weak correlation with other independent variables.

```r
#Removing outliers
#Cooks distance measurement
options(repr.plot.width=10, repr.plot.height=5)
cooksd <- cooks.distance(model1)
plot(cooksd, pch="*", cex=2, main="Influential Obervations by Cooks distance")
abline(h = 5*mean(cooksd, na.rm=T), col="red")
```

```r
#Remove outliers
ctdata <- cdata[(cdata$county != cdata[influential[1],]$county) & (cdata$county != cdata[influential[2],]$county),]
```

```r
#Create model2
options(repr.plot.width=10, repr.plot.height=5)
model2 <- lm(crmrte ~ prbarr + prbconv + density + taxpc + wage + pctymle + polpc + pctmin80 + mix, data = ctdata)
coeftest(model2, vcov = vcovHC)
plot(model2, which = 1)
plot(model2, which = 5)
```

#### 3.3 Model 3 - Good Fit with Parsimony

Adjustments after Model2 results - 
1. Removing statistically insignificant variables, 'mix', 'taxpc' and 'wage'.

#### 3.5 Causal Estimation

The model3 shows strong alignment with required OLS assumptions. With Zero conditional mean, exogeneity and very low residuals, the model3 seems to showing a good indication for causal estimation. However, the apparent causation could well be a lot weaker for the following reasons:
- With more data, the model robustness could easily go down. Currently, a relatively low number of observations (~90) are used to create the models.
- Biased sampling methods and poor data collection strategies could have a negative impact on the estimation.
- In a practical setting, it is very hard to make sure models contain all of the data points relevant to the thing that we are attempting to model.  This leads to most models having some level of omitted variable bias. 

Finally, even if the causal estimation is hard to come by for this analysis, we can confidently conclude this is an associative model.

#### 3.6 Omitted Variable Effects

The models built in this report are quite powerful considering the smaller data set and limited variables.  They do not explain everything however, and are influenced by many factors not immediately obvious from the models and their analysis.  The are many omitted variables that have effects on this model, and that we would like to have if we were to improve these models substantially.

Unemployment for example has a large effect on the crime rate in an area.  A higher unemployment rate generally means more crime.  Some of this effect may be shown in the wage variables, as high unemployment often drives wages down.  This means that the lack of unemployment data may be moving the wage coefficients downwards as the unemployment effect would be swallowed into the wage effect.  If the unemployment statistic was present it would likely by a significant negative bias.

While we do have the percentage of young males in an area, they are not the only demographic that causes crime.  A more granular breakdown of ages/genders, would be extremely useful.  The omission of this, causes the models to be agnostic of the effects of female / older populations.  They are likely small negative biases, higher rates of females and older populations likely correlate to lower crime rates.

Another adaptation of an existing statistic is the funding of the police in a county.  The police per capita will cover some of the same things, but the budget that the police have to work with is also very important.  Police budget is likely tied closely with police per capita and density.  Its omission likely introduces bias that moves their coefficients positively away from 0.

Another interesting omitted variable to look at is the politcal or religous leanings of an area.  It is very difficult to say what bias, if any, their omission introduces, but it would certainly be interesting to look at.

Other factors tied into the minority variable would be good to know.  The percent minority variable has strong relationships with the crime rate, but it is very likely that more is at play than just their status as minorities.  The omission of variables such as homelessness, drug use, and social bias that disproportionately effect minorities has a definite bias on the effect of minority percentage in the model.  These things likely cause a positive bias away from 0 on the minority coefficient, inflating the percieved importance of the minority variable.  It is difficult to say how much of the key effect of the minority percentage is due to these omitted variables vs how much is actually due to the minority status.  However, it is safe to say that at least some of the effect seen as the minority percent is due to these omitted variables.

### 4. Major Takeaways and Proposed Policy Changes

While these models do a good job of modeling the crime rate based on a handful of statistics, it is important to keep in mind that these numbers are not perfect nor do they show causality.  For instance a change in the probability of arrest may be corellated with a change in crime rate, it does not mean that a unit change in the probability of arrest will cause a corresponding change in the crime rate.

With that being said, these models can tell a lot about what changes could be made that might help decrease the crime rate.  Across the models several variables stand out, specifically the probability of arrest, probability of conviction, minority percentage, and the density of a county.

The probability of arrest and probability of conviction are things that can be directly controlled by politicians and the communities at large.  They both fall under the category of certainty of punishment, which means that the correlation between increases in these and decreases in the crime rate are not related to how severe the punishment is.  Taking this into consideration, our team would suggest a policy change resulting in a lower tolerance level for criminal activities.  This would mean that more crimes would lead to arrests and convictions.  However, we would also suggest that any policy keep in mind the human factor when considering the severity of the new punishments.  These models did not show a strong relationship between the severity of a punishment in terms of the probability of prison or the average sentence and the crime rate.  An increase in the probability of punisment will likely go much futher in reducing the crime rate compared to an increase in the severity of punishment.

This report shows that there is a relationship between the percentage of minorities in a county and the crime rate in the county.  This can be a more complicated variable to discuss in terms of pure statistics for a number of reasons.  Minorities are disproportionately affected by poverty, can be targeted, and generally live in high density areas.  The effect of the minority percentage is fairly large, and not fully explored in this report.  Our team would suggest funding a more focused statistical analysis on this relationship before making policy changes.

Another relationship found in this report is the corellation of crime and the denisty of a county.  This seems obvious, more people means more opportunity for crime, however it may not be so simple. ### Appendix -  A
Helper functions

```r
getLogTran <- function(df, cols, tranType, tranTypeDep){
    corList <- list()
    
    for(coln in cols){
               
        if(tranType == ''){
          corList <- append(corList, cor(df[[coln]], tranVar(df$crmrte, tranTypeDep)))   
            
        }else{
            
          newColN <- paste(coln, tranType, sep="_")
          df[[newColN]] <- tranVar(df[[coln]], tranType)    
          corList <- append(corList, cor(df[[newColN]], tranVar(df$crmrte, tranTypeDep)))              
      }        
        
    }
    
    return (corList)
}
    
tranVar <- function(var1, tranType){
    if(tranType == 'lg'){
        return (log(var1))
    } else if(tranType == 'sq'){
        return (var1^2)
    } else if(tranType == 'sqt'){
        return (sqrt(var1))
    } else {
        return (var1)
    }
}
```
